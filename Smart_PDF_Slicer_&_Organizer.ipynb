```
!pip install PyMuPDF

import os

import fitz  # PyMuPDF

from google.colab import drive

from google.colab import files



# 1. Mount Google Drive

print("Mounting Google Drive...")

drive.mount('/content/drive')



def process_pdf_to_sessions():

    print("\n" + "="*40)

    # 2. Upload the PDF file

    print("üìÇ Please upload the PDF file here:")

    uploaded = files.upload()



    if not uploaded:

        return print("‚ö†Ô∏è No file uploaded. Operation cancelled.")



    # Get the uploaded file name

    pdf_path = list(uploaded.keys())[0]



    # 3. Organize Sessions (Grade & Number of Sessions)

    print("\n" + "="*40)

    grade_name = input("üìù What is the Grade name? (e.g., Grade 5): ").strip()

    if not grade_name:

        return print("‚ö†Ô∏è Invalid name. Operation cancelled.")



    try:

        num_sessions = int(input("üî¢ How many sessions do you need? (Enter a number): "))

    except ValueError:

        return print("‚ö†Ô∏è You must enter a valid number! Operation cancelled.")



    # Create the main folder in Google Drive

    main_folder_path = f"/content/drive/MyDrive/{grade_name}"

    os.makedirs(main_folder_path, exist_ok=True)

    print(f"\n‚úÖ Main folder created in Drive: {grade_name}")



    # 4. Open the PDF to get the total number of pages

    try:

        doc = fitz.open(pdf_path)

        total_pages = len(doc)

        print(f"üìÑ PDF loaded successfully. Total pages: {total_pages}.")

    except Exception as e:

        return print(f"‚ö†Ô∏è Error opening PDF file: {e}")



    print("\n" + "="*40)



    # 5. Ask for page ranges for each session and extract images

    for i in range(1, num_sessions + 1):

        session_name = f"Session {i}"

        session_path = os.path.join(main_folder_path, session_name)

        os.makedirs(session_path, exist_ok=True)



        print(f"\n--- Preparing {session_name} ---")



        # Validate page number inputs

        while True:

            try:

                start_page = int(input(f"‚¨ÖÔ∏è Start page? (1 to {total_pages}): "))

                end_page = int(input(f"‚û°Ô∏è End page? (1 to {total_pages}): "))



                if 1 <= start_page <= end_page <= total_pages:

                    break

                else:

                    print("‚ö†Ô∏è Invalid range! Make sure start page is <= end page, and within the document limits.")

            except ValueError:

                print("‚ö†Ô∏è Please enter numbers only.")



        print(f"‚è≥ Extracting pages {start_page} to {end_page} as high-quality images...")



        # Extract images (subtract 1 because Python index starts at 0)

        count = 0

        for page_num in range(start_page - 1, end_page):

            page = doc[page_num]



            # High-quality extraction (Zoom 2.0)

            zoom = 2.0

            mat = fitz.Matrix(zoom, zoom)

            pix = page.get_pixmap(matrix=mat)



            # Save the image directly to the session folder in Google Drive

            image_filename = f"{session_name}_page_{page_num + 1}.png"

            image_path = os.path.join(session_path, image_filename)

            pix.save(image_path)

            count += 1



        print(f"‚úÖ Successfully saved {count} images in the '{session_name}' folder.")



    doc.close()

    print("\n" + "="*40)

    print(f"üéâ All done! The PDF has been sliced and organized inside the '{grade_name}' folder in your Google Drive.")



# Run the tool

process_pdf_to_sessions()
```
